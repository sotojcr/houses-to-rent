```python
import csv
```

```python
arquivo = open('contatos.csv', 'r', encoding='utf8')

planilha = list(csv.reader(arquivo, delimiter=';', lineterminator='\n'))
```

```python
def eh_nome(arquivo):
    nomes = []
    for linha in planilha:
        for elemento in linha:
            if elemento.isalpha() == True:
                nomes.append(elemento)
    return nomes

def eh_email(arquivo):
    emails = []
    for linha in planilha:
        for elemento in linha:
            if '@' in elemento:
                emails.append(elemento)
    return emails

def eh_telefone(arquivo):
    telefones = []
    for linha in planilha:
        for elemento in linha:
            if elemento.isnumeric() == True:
                telefones.append(elemento)
    return telefones
```

```python
telefones = eh_telefone(planilha)
nomes = eh_nome(planilha)
emails = eh_email(planilha)

dados_estruturados = []

for i in range(len(telefones)):
    dados_estruturados.append([nomes[i], telefones[i], emails[i]])
    
    
print(dados_estruturados)
```

```python
class Interface:
    def __init__(self, fornecedores):
        self.fornecedor = fornecedores
        
    def insere_fornecedor(self):
        nome = input('Informe o nome do fornecedor: ')
        telefone = input('Informe o telefone do fornecedor: ')
        email = input('Informe o e-mail do fornecedor: ')
        
        self.fornecedor.append([nome, telefone, email])
        
    def remove_fornecedor(self):
        informacao = input('Digite alguma informação do fornecedor (nome/telefone/email): ')
        verifica = False
        
        for linha in self.fornecedor:
            for valor in linha:
                if informacao == valor:
                    self.fornecedor.remove(linha)
                    verifica = True
                    print('Fornecedor removido!')
        if verifica == False:
            print('Fornecedor não encontrado')
                                      
    def busca_fornecedor(self):
        informacao = input('Digite alguma informação do fornecedor (nome/telefone/email): ')
        verifica = False
        
        for linha in self.fornecedor:    
            if informacao in linha:
                print(linha)
                verifica = True
        if verifica == False:
            print('Fornecedor não encontrado')
    
    def transforma_em_csv(self):
        arquivo = open('fornecedores.csv', 'w')
        
        escritor = csv.writer(arquivo, delimiter=';', lineterminator='\n')
        escritor.writerows(self.fornecedor)
        
        arquivo.close()
        print('Arquivo transformado com sucesso')
    
    def run(self):
        print('1- Inserir novo fornecedor\n\
        2- Remover fornecedor\n\
        3- Busca fornecedor\n\
        4- Transformar em arquivo .csv\n\
        0- Sair')
        
        escolha = input('O que fazer: ')
        
        while escolha != '1' and escolha != '2' and escolha != '3' and escolha != '4' and escolha != '0':
            print('Opção inválida')
            escolha = input('O que fazer: ')
            
        while escolha != '0':   
            if escolha == '1':
                self.insere_fornecedor()
                escolha = input('O que fazer: ')
            elif escolha == '2':
                self.remove_fornecedor()
                escolha = input('O que fazer: ')
            elif escolha == '3':
                self.busca_fornecedor()
                escolha = input('O que fazer: ')
            elif escolha == '4':
                self.transforma_em_csv()
                escolha = input('O que fazer: ')
        
        print('===FINALIZADO===')
            
```

```python
sistema = Interface(dados_estruturados)
```

```python
sistema.run()
```

```python
class Agenda:
    def __init__(self):
        self.fornecedores_cadastrados = {}
        self.novos_fornecedores = []
#Classe que possui os locais de armazenamento das informações dos fornecedores e dos objetos gerados
        
        
class Fornecedor:
    def __init__(self, identificador, nome, telefone, email):
        self.nome = nome
        self.telefone = telefone
        self.email = email                
        self.identificador = identificador
        
```

```python
class SistemaCadastros:
    max_cadastro = 0 # Conta o números de fornecedores na agenda
    identificador = 1 # Número de identificação (não regride após algum contato ser removido)
    
    def __init__(self):
        self.lista_fornecedores = dados_estruturados
        self.agenda = Agenda()
    
    def popula_agenda(self):
        for i in range(1, len(self.lista_fornecedores) + 1):
            self.agenda.fornecedores_cadastrados[i] = self.lista_fornecedores[i-1]
            SistemaCadastros.max_cadastro += 1
            SistemaCadastros.identificador += 1
    #Método para adicionar os fornecedores já existentes na nova agenda    
    
    def visualizar_agenda(self):# Visualiza a agenda completa 
        for chave, valor in self.agenda.fornecedores_cadastrados.items():
            print(chave, valor)
    
    @staticmethod        
    def adiciona_ao_grupo(nome_grupo, fornecedor, agenda):
        
        if not nome_grupo in agenda.keys():
            agenda[nome_grupo] = {fornecedor.identificador : [fornecedor.nome, fornecedor.telefone, fornecedor.email]}
        
            SistemaCadastros.max_cadastro += 1
            SistemaCadastros.identificador += 1
        else:
            agenda[nome_grupo][fornecedor.identificador] = [fornecedor.nome, fornecedor.telefone, fornecedor.email]
        
            SistemaCadastros.max_cadastro += 1
            SistemaCadastros.identificador += 1

    # Método estático para auxiliar o método cadastra_fornecedor a adicionar fornecedores a grupos
            
    
    def cadastra_fornecedor(self):
       
        if SistemaCadastros.max_cadastro < 75:   
            
            identificador = SistemaCadastros.identificador
            nome = input('Informe o nome do novo fornecedor: ')
            
            n_contatos = int(input('quantos numeros de telefone o fornecedor possui: '))
            telefones = []
            
            for i in range(n_contatos):
                ddd = input('Informe o DDD: ')
                numero = input('Informe o telefone do fornecedor: ')
                telefone = ddd + numero
                
                while (telefone.isnumeric() != True) and (len(telefone) == 10 or len(telefone) == 11):
                
                    print('Telefone inválido')

                    ddd = input('Informe o DDD: ')
                    numero = input('Informe o telefone do fornecedor: ')
                    telefone = ddd + numero
                
                telefones.append(telefone)
            
            n_emails = int(input('Informe a quantidade de e-mails que o fornecedor possui: '))
            emails = []
            
            for i in range(n_emails):
                email = input('Informe o e-mail do fornecedor: ')

                while not ('.' in email and '@' in email):
                    print('E-mail inválido, tente novamente')
                    email = input('Informe o e-mail do fornecedor: ')
                emails.append(email)
            
            novo_fornecedor = Fornecedor(identificador, nome, telefones, emails)
            self.agenda.novos_fornecedores.append(novo_fornecedor)
            
            print('1- Sim\n2- Não')
            opcao_usuario = input('Deseja adicionar o fornecedor a um grupo: ')
            
            while opcao_usuario != '1' and opcao_usuario != '2':
                print('Opção inválida')
                opcao_usuario = input('Deseja adicionar o fornecedor a um grupo: ')
                
            if opcao_usuario == '1':
                nome_grupo = input('Informe o nome do grupo: ')
                SistemaCadastros.adiciona_ao_grupo(nome_grupo, novo_fornecedor, self.agenda.fornecedores_cadastrados)
            
            elif opcao_usuario == '2':
                self.agenda.fornecedores_cadastrados[novo_fornecedor.identificador] = [novo_fornecedor.nome,
                                                                                       novo_fornecedor.telefone,
                                                                                       novo_fornecedor.email]
            else:
                print('Opção inválida')
                opcao_usuario = input('Deseja adicionar o fornecedor a um grupo: ')
        else:
            print('Agenda cheia')
            
    @staticmethod
    def altera_telefone_email(agenda, id_fornecedor, index):
        if id_fornecedor not in agenda:

            for id_cadastrado, informacoes in agenda.items():
                if type(informacoes) == dict and id_fornecedor in informacoes:
                    total_infos_armaz = len(informacoes[id_fornecedor][index]) + 1
                    if len(informacoes[id_fornecedor][index]) > 1:
                        informacao_substituida = int(input('Qual telefone deseja alterar?'
                                                     f'{list(range(1, total_infos_armaz))}'))

                        informacao_substituida -= 1  # Ajusta ao index da lista de telefones
                        informacoes[id_fornecedor][index][informacao_substituida] = \
                            input('Digite a nova informação de cadastro: ')
                        print('Informação alterada')
                    else:
                        informacoes[id_fornecedor][index][0] = input('Digite a nova informação de cadastro: ')
                        print('Informação alterada')
        else:
            if type(agenda[id_fornecedor][index]) == list:
                informacao_substituida = int(input('Qual das informações a seguir deseja alterar? '
                                             f'{list(range(1, len(agenda[id_fornecedor][index]) + 1))}'))
                informacao_substituida -= 1
                agenda[id_fornecedor][index][informacao_substituida] = \
                    input('Digite a nova informação de cadastro: ')
            else:
                agenda[id_fornecedor][index] = input('Digite a nova informação de cadastro: ')

    def altera_cadastro(self):
        print('1- Alterar nome'
              '\n2- Alterar telefone'
              '\n3- Alterar email')

        escolha = input('Qual informação deseja alterar: ')

        while escolha != '1' and escolha != '2' and escolha != '3':
            print('Opção inválida')
            escolha = input('Qual informação deseja alterar: ')

        id_fornecedor = input('Informe o ID do usuário que deseja adicionar em um grupo: ')
        id_fornecedor = int(id_fornecedor)

        if escolha == '1':
            try:
                for chave, valor in self.agenda.fornecedores_cadastrados.items():
                    if id_fornecedor == chave:
                        valor[0] = input('Informe o nome do fornecedor: ')
                        print('Nome alterado')
                    elif type(valor) == dict and id_fornecedor in valor:
                        valor[id_fornecedor][0] = input('Informe o nome do fornecedor: ')
                        print('Nome alterado')
            except:
                print('Fornecedor não encontrado')
        elif escolha == '2':
            SistemaCadastros.altera_telefone_email(self.agenda.fornecedores_cadastrados, id_fornecedor, 2)
            
    @staticmethod
    def remove_do_grupo(agenda, id_fornecedor):#Auxilia na alteração de um fornecedor de um grupo para outro
        
        for chave, valor in agenda.items():
            if type(valor) == dict and id_fornecedor in valor:
                break

        return agenda[chave].pop(id_fornecedor)

    
    def altera_grupo(self):
        id_fornecedor = int(input('Informe o ID do usuário que deseja adicionar em um grupo: '))
        nome_grupo = input('Informe o nome do grupo: ')
        
        try:
            if not id_fornecedor in self.agenda.fornecedores_cadastrados.keys():
                infos = SistemaCadastros.remove_do_grupo(self.agenda.fornecedores_cadastrados, id_fornecedor)

                if not nome_grupo in self.agenda.fornecedores_cadastrados.keys():
                    self.agenda.fornecedores_cadastrados[nome_grupo] = {id_fornecedor:infos}
                else:
                    self.agenda.fornecedores_cadastrados[nome_grupo][id_fornecedor] = infos
            
            else:
                infos = self.agenda.fornecedores_cadastrados.pop(id_fornecedor)

                if not nome_grupo in self.agenda.fornecedores_cadastrados.keys():
                    self.agenda.fornecedores_cadastrados[nome_grupo] = {id_fornecedor:infos}
                else:
                    self.agenda.fornecedores_cadastrados[nome_grupo][id_fornecedor] = infos
        except:
            print('Fornecedor não encontrado')
    
    def remove_fornecedor(self):
        id_fornecedor = int(input('Entre com o ID do fornecedor que deseja remover: '))
        try:
            if id_fornecedor in self.agenda.fornecedores_cadastrados.keys():
                self.agenda.fornecedores_cadastrados.pop(id_fornecedor)
                print('fornecedor removido')

            else:
                SistemaCadastros.remove_do_grupo(self.agenda.fornecedores_cadastrados, id_fornecedor)
        except:
            print('Fornecedor não encontrado')
            
    def remove_grupo(self):
        nome_grupo = input('Entre com o nome do grupo que deseja remover: ')
        
        try:
            self.agenda.fornecedores_cadastrados.pop(nome_grupo)
        except:
            print('Grupo não encontrado')
    
    def buscar_contato(self):
        id_fornecedor = int(input('Informe o ID do fornecedor: '))
        try:
            if not id_fornecedor in self.agenda.fornecedores_cadastrados.keys():
                print(SistemaCadastros.remove_do_grupo(self.agenda.fornecedores_cadastrados, id_fornecedor))
            else:
                print(self.agenda.fornecedores_cadastrados[id_fornecedor])
        except:
            print('Usuário não encontrado')
        
    
    def transforma_csv(self):
        
        #Gera uma lista de listas para ser transformado em csv
        
        dados_organizados = []

        for valor in self.agenda.fornecedores_cadastrados.values():
            dados = []
            
            if type(valor) == list:
                for informacao in valor:
                    #Caso haja mais de um telefone ou email o algaritmo manda cada um deles para uma nova lista
                    if type(informacao) == list:
                        for dado in informacao:
                            dados.append(dado)
                    #se não, ele apenas coloca o dado na nova lista
                    else:
                        dados.append(informacao)
                dados_organizados.append(dados)
            #Realiza o mesmo processo do if, porém, para os grupos criados pelo usuário 
            else:
                for fornecedor in valor.values():
                    dados = []
                    for informacao in fornecedor:
                        if type(informacao) == list:
                            for dado in informacao:
                                dados.append(dado)
                        else:
                            dados.append(informacao)
                    dados_organizados.append(dados)
        
        # Transforma em csv
        arquivo = open('fornecedores.csv', 'w')
        
        escritor = csv.writer(arquivo, delimiter=';', lineterminator='\n')
        
        escritor.writerows(dados_organizados)
        arquivo.close()
    
    def busca_grupo(self):
        print('Esta opção removerá o grupo e tudo que houver dentro dele')
        
        nome_grupo = input('Informe o nome do grupo (0 para cancelar): ')
        
        try:
            print(self.agenda.fornecedores_cadastrados[nome_grupo])
        except:
            print('Grupo não encontrado')
                
    def run(self):
        self.popula_agenda()
        
        print('1- Visualizar agenda completa\n\
        2- Cadastrar fornecedor\n\
        3- Alterar/adicionar fornecedor a grupo(existente ou não)\n\
        4- Remover Fornecedor\n\
        5- Remover Grupo\n\
        6- Busca contato\n\
        7- Busca Grupo\n\
        0 - Sair')
        
        escolha = input('Escolha a opção de navegação: ')
        
        while escolha != '0' and escolha != '1' and escolha != '2' and escolha != '3' and escolha != '4' and escolha != '5' and escolha != '6' and escolha != '7':
            print('Opção inválida')
            escolha = input('Escolha a opção de navegação: ')
        
        while escolha != '0':
            if escolha == '1':
                self.visualizar_agenda()
                escolha = input('Escolha a opção de navegação: ')
            elif escolha == '2':
                self.cadastra_fornecedor()
                escolha = input('Escolha a opção de navegação: ')
            elif escolha == '3':
                self.altera_grupo()
                escolha = input('Escolha a opção de navegação: ')
            elif escolha == '4':
                self.remove_fornecedor()
                escolha = input('Escolha a opção de navegação: ')
            elif escolha == '5':
                self.remove_grupo()
                escolha = input('Escolha a opção de navegação: ')
            elif escolha == '6':
                self.buscar_contato()
                escolha = input('Escolha a opção de navegação: ')
            elif escolha == '7':
                self.busca_grupo()
                escolha = input('Escolha a opção de navegação: ')
            else:
                break
        print('===FINALIZADO===')
    
    
```
